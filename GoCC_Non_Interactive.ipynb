# Key Concepts (Goal Models)



### Actors

> **Actor**  a person, role, or organization that has intentions and participates in the system.

---

### Intentional Elements

> **Goal**  something clear-cut that an actor wants to achieve.

> **Task**  an action or step that an actor wants or must perform.

> **Resource**  a thing (physical, informational, or skill-based) that an actor needs to possess, usually to perform a task.

> **Soft Goal**  a goal an actor wants to achieve but with a fuzzy or non-binary satisfaction criterion.  
> Typically used to represent qualities or decision criteria.


---

### Links

> **AND Decomposition**  to achieve the **parent**, **all children** must be achieved.

> **OR Decomposition (MeansEnds)**  to achieve the **parent**, **one or more alternatives** (children) must be achieved.

> **Dependency**  an **actor** depends on something from **another actor**, usually to achieve a **goal** or perform a **task**.  

> **Contribution**  an element in an actor has a **positive or negative effect** on a **soft goal**.  The effect can be **strongly positive (MAKES)**, or **strongly negative (BREAKS)**.

---
### Nodes

> **Leaf Node** are defined as nodes with NO Incoming decomposition (AND/OR [Means-End]) or contribution links (Makes, Breaks, Helps, Hurts) and  NO Outgoing dependency links (Depends).


> **Root (Top) Nodes** are defined as nodes with NO Outgoing decomposition (AND/OR [Means-End]) or contribution links (Makes, Breaks, Helps, Hurts) and NO Incoming dependency links (Depends).


> **Transitive-children:** all nodes reachable downstream from Destination by repeatedly following contribution links of the specified type


---




# Satisfaction an propagation of intentional elements

---
## Decomposition Links

---
##### AND Decomposition

| Child 1 | Child 2 | Parent |
|--------|--------|--------|
| Satisfied (, ) | Satisfied (, ) | Satisfied (, ) |
| Satisfied (, ) | Pending (, ) | Pending (, ) |
| Pending (, ) | Satisfied (, ) | Pending (, ) |
| Pending (, ) | Pending (, ) | Pending (, ) |
| Unknown | Unknown | Unknown |
| Unknown | Pending (, ) | Pending (, ) |
| Pending (, ) | Unknown | Pending (, )  |
| Unknown | Satisfied (, ) | Unknown |
| Satisfied (, ) | Unknown | Unknown |
   

---
##### OR Decomposition

| Child 1 | Child 2 | Parent |
|--------|--------|--------|
| Satisfied (, ) | Satisfied (, ) | Satisfied (, ) |
| Satisfied (, ) | Pending (, ) | Satisfied (, ) |
| Pending (, ) | Satisfied (, ) |Satisfied (, ) |
| Pending (, ) | Pending (, ) | Pending (, ) |
| Unknown | Unknown | Unknown |
| Unknown | Pending (, ) | Pending (, )|
| Pending (, ) | Unknown | Pending (, )|
| Unknown | Satisfied (, ) |  Satisfied (, )  |
| Satisfied (, ) | Unknown |  Satisfied (, ) |

---

## Contribution Links
---


> **Contribution propagation**  contribution links propagate according to the **type of contribution link**.  For example, a **Source MAKES Destination** propagates a strongly positive effect from Source to Destination.


We distinguish two cases:

- **Case A (Destination = Unknown):** the link can *set* the Destination.
- **Case B (Destination  Unknown):** the link may *force Pending* on parts of the downstream chain (your transitive-children rule).

---

#### Case A  Destination is **Unknown**

| Link Type | Source value | New Destination value |
|----------|--------------|-----------------------|
| MAKE  | Unknown | Unknown |
| MAKE  | Satisfied $(\top,\bot)$ | Satisfied $(\top)$ |
| MAKE  | Pending $(\top,\top)$ | Pending $(\top,\top)$ |
| BREAK | Unknown | Unknown |
| BREAK | Satisfied $(\top,\bot)$ | Denied $(\bot)$ |
| BREAK | Pending $(\top,\top)$ | Pending $(\top,\top)$ |

---

#### Case B  Destination already has a value (**Destination  Unknown**)

| Link Type | If Source is | Destination becomes | Extra rule (propagation to transitive-children) |
|----------|----------------|----------------------|-----------------------------------------------|
| MAKE  | Satisfied $(\top,\bot)$ | Satisfied $(\top)$ | Set **Pending $(\top,\top)$** for all Destination **transitive-children** along a **BREAK** chain |
| BREAK | Satisfied $(\top,\bot)$ | Denied $(\bot)$ | Set **Pending $(\top,\top)$** for all Destination **transitive-children** along a **MAKE** chain |
| MAKE / BREAK | Unknown | (no change) | none |
| MAKE / BREAK | Pending $(\top,\top)$ | Pending $(\top,\top)$ | none |

---

## Dependency Links

---

> **Dependency propagation**  dependency links propagate values in the **opposite direction** of the `depends` arrow.  
> If the **Destination depends on the Source**, then the label/value at the **Source** is propagated directly to the **Destination**.

### Dependency Propagation

| Source value | Dependum value | Destination value |
|--------------|----------------|-------------------|
| Unknown | Unknown | Unknown |
| Satisfied $(\top,\bot)$ | Satisfied $(\top,\bot)$ | Satisfied $(\top,\bot)$ |
| Pending $(\top,\top)$ | Pending $(\top,\top)$ | Pending $(\top,\top)$ |

# Visualizing Alignments

```python
# Formateo de alignments viaulization

def mv_to_str(mv):
    """Convert PM4Py move objects to readable strings."""
    if mv == ">>":
        return ">>"
    if mv is None:
        return ""
    # Transition object
    if hasattr(mv, "label"):
        return mv.label if mv.label is not None else ""
    s = str(mv)
    return "" if s in ("None", "") else s

def is_tau(mv):
    """True if mv is a silent move (None label / None / 'None' / '')."""
    if mv is None:
        return True
    if hasattr(mv, "label") and mv.label is None:
        return True
    s = str(mv)
    return s in ("None", "")

def format_alignment_two_rows_clean(aln_dict, sep=" | ", left_pad=2, hide_tau_only=True):
    pairs = aln_dict.get("alignment", [])

    filtered = []
    for (log_mv, model_mv) in pairs:
        # Recommended: hide only pure silent-routing columns (>> / )
        if hide_tau_only:
            if log_mv == ">>" and is_tau(model_mv):
                continue
        else:
            # Hide any column where model is 
            if is_tau(model_mv):
                continue
        filtered.append((log_mv, model_mv))

    log_cells = [mv_to_str(l) for (l, m) in filtered]
    model_cells = [mv_to_str(m) for (l, m) in filtered]

    widths = [max(len(log_cells[i]), len(model_cells[i])) for i in range(len(filtered))]
    log_cells = [log_cells[i].ljust(widths[i]) for i in range(len(filtered))]
    model_cells = [model_cells[i].ljust(widths[i]) for i in range(len(filtered))]

    log_prefix = "LOG   :".ljust(left_pad + 7)
    model_prefix = "MODEL :".ljust(left_pad + 7)

    log_line = (" " * left_pad) + log_prefix + sep.join(log_cells)
    model_line = (" " * left_pad) + model_prefix + sep.join(model_cells)

    ruler_len = max(len(log_line), len(model_line))
    ruler_line = (" " * left_pad) + "-" * (ruler_len - left_pad)

    return log_line, model_line, ruler_line

def print_alignment_clean(aln_dict, label="", sep=" | ", hide_tau_only=True):
    log_line, model_line, ruler = format_alignment_two_rows_clean(
        aln_dict, sep=sep, hide_tau_only=hide_tau_only
    )
    print(f"{label} =")
    print(log_line)
    print(model_line)
    print(ruler)
```
